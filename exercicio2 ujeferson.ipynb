Crie um programa que peça um número e mostre:
▶ Número positivo
▶ Número negativo
▶ Número igual a zero
Dica: Use as estruturas if, elif e else para resolver
esse problema.

```python
numero = int(input("Digite um número: "))
if numero > 0:
    print("Número positivo")
elif numero < 0:
    print("Número negativo")
else:
    print("Número igual a zero")
```
